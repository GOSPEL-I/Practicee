```python
account_balance = 1000.00

def check_balance():
    print(f"\nYour current balance is: #{account_balance:.2f}")

def deposit():
    global account_balance
    while True:
        try:
            amount = float(input("Enter the amount to deposit: "))
            if amount <= 0:
                print("Deposit amount must be positive.")
            else:
                account_balance += amount
                print(f"Successfully deposited #{amount:.2f}. Your new balance is: #{account_balance:.2f}")
                break
        except ValueError:
            print("Invalid input. Please enter a numerical amount.")

def withdraw():
    global account_balance
    while True:
        try:
            amount = float(input("Enter the amount to withdraw: "))
            if amount <= 0:
                print("Withdrawal amount must be positive.")
            elif amount > account_balance:
                print(f"Insufficient funds. Your current balance is: #{account_balance:.2f}")
            else:
                account_balance -= amount
                print(f"Successfully withdrew #{amount:.2f}. Your new balance is: #{account_balance:.2f}")
                break
        except ValueError:
            print("Invalid input. Please enter a numerical amount.")

def display_menu():
    print("\nBanking Menu:")
    print("1: Check Balance")
    print("2: Deposit")
    print("3: Withdraw")
    print("4: Exit")


def run_banking_system():
    while True:
        display_menu()
        try:
            choice = input("Choose an option (1-4): ")
            if choice == '1':
                check_balance()
            elif choice == '2':
                deposit()
            elif choice == '3':
                withdraw()
            elif choice == '4':
                print("Thank you for using pora banking system.")
                break
            else:
                print("Invalid option. Please choose a number between 1 and 4.")
        except Exception as e:
            print(f"An unexpected error occurred: {e}")

run_banking_system()
```
